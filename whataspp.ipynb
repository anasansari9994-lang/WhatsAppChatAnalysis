```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
import re
```

```python
chat = open(r"C:\Users\ANAS ANSARI\OneDrive\Desktop\whatsapp chat analysis\chatgroup1\WhatsApp Chat with SME OCT DEC 2025 AIILSG Batch.txt" , 'r' , encoding='utf-8')
```

```python
data = chat.read()
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:am|pm|AM|PM)\s-\s'
pattern_f = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?(?:am|pm|AM|PM)'

```

```python
message = re.split(pattern , data)[1:]
len(message)
```

```python
dates = re.findall(pattern_f , data)
len(dates)
```

```python
df = pd.DataFrame({'user_message' :message , 'message_date' :dates})
df['message_date'] = pd.to_datetime(df['message_date'], format='%d/%m/%Y, %I:%M %p')
print(df.head())
```

```python
df.rename(columns={'message_date' : 'date'} , inplace=True)
df.head
```

```python
df[['user' , 'message']] = (
    df['user_message']
    .str.split(':' , n=1 , expand=True)
    .apply(lambda x : x.str.strip()))
```

```python
df.drop(columns=['user_message'] , inplace=True)
df.tail()
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```

```python
df.tail()
```

```python
df.drop(columns=['date'] , inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df_grouped = df.groupby(['year','month']).count()['message'].reset_index()
df_grouped
df_grouped['date'] = pd.to_datetime(
    df_grouped['year'].astype(str) + '-' + df_grouped['month'].astype(str)
)
df_grouped.head()
```

```python
df.head()
df_mostactivetime = df.groupby(['hour']).count()['message'].reset_index()
df_mostactivetime.rename(columns={'message' : 'message Count'} , inplace = True)
df_mostactivetime
```

```python
import plotly.express as px

fig = px.line(
    df_grouped,
    x='date',
    y='message',
    title='Monthly Chat Activity',
    markers=True
)

fig.update_layout(
    xaxis_title='Time',
    yaxis_title='Number of Messages'
)

fig.show()
```
