```python
import torch
from torch.utils.data import Dataset

class SentimentDataset(Dataset):
    def __init__(self, encodings, labels):
        self.encodings = encodings
        self.labels = labels

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        item = {key: val[idx] for key, val in self.encodings.items()}
        item['labels'] = torch.tensor(self.labels[idx], dtype=torch.long)
        return item
```

```python
from transformers import AutoTokenizer , AutoModelForSequenceClassification
import torch.nn as nn
class HybridToxicitymodel(nn.Module):
    def __init__(self , model_a , model_b):
        super().__init__()

        self.multilingualModel = AutoModelForSequenceClassification.from_pretrained(model_a)
        self.HinglishModel = AutoModelForSequenceClassification.from_pretrained(model_b)

        self.model_a_tokenizer = AutoTokenizer.from_pretrained(model_a)
        self.model_b_tokenizer = AutoTokenizer.from_pretrained(model_b)


        self.multilingualModel.eval()
        self.HinglishModel.eval()

    def forward(self, txt):

        input_a = self.model_a_tokenizer(txt, padding=True , truncation=True , return_tensors='pt ' , max_length = 128)
        input_b = self.model_b_tokenizer(txt, padding=True , truncation=True , return_tensors='pt' , max_length = 128)

        with torch.no_grad():


            output_a = self.multilingualModel(input_ids_a = input_a['input_ids'] , attention_mask_a = input_a['attention_mask'])
            output_b = self.HinglishModel(input_ids_b = input_b['input_ids'] , attention_mask_b = input_b['attention_mask'])

            logits_a = output_a.logits
            logits_b = output_b.logits

            
            final_logits = (logits_a + logits_b) / 2

            probs = torch.softmax(final_logits, dim=1)

            predicted_labels = torch.argmax(probs, dim=1)
            confidence_scores = torch.max(probs, dim=1).values
            
            return {
                "final_logits": final_logits,
                "probs": probs,
                "predicted_labels": predicted_labels,
                "confidence_scores": confidence_scores
            }
```

```python
import re
def normalize_wording(text):
    text = text.lower()

    text = re.sub(r'\s+', ' ', text).strip()
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'@\w+', '', text)
    return text
```

```python
df = df[df['text'].str.len() > 10]
df = df[df['text'].str.len() < 300]
df.shape
```

```python
from datasets import load_dataset
from datasets import Dataset
dataset = Dataset.from_pandas(df)
dataset = dataset.map(
    lambda x: {'text': normalize_wording(str(x['text'])), 'label': x['label']},
    remove_columns=dataset.column_names 
)
```

```python
test_data = load_dataset('csv' , data_files = {"test": "test_dataset.csv"})
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

model_name = "xlm-roberta-base"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name, num_labels=3 , hidden_dropout_prob=0.3, attention_probs_dropout_prob=0.3)
```

```python
from datasets import DatasetDict

# Split only train
train_test_split = dataset.train_test_split(test_size=0.1, seed=42)
train_val_split = DatasetDict({
    "train": train_test_split["train"],
    "validation": train_test_split["test"]
})
```

```python


def tokenization(examples):
    return tokenizer(examples['text'], truncation=True, padding="max_length", max_length=256)

train_val_split = train_val_split.map(tokenization, batched=True)

# 3. Finalize columns
train_val_split = train_val_split.rename_column("label", "labels")
train_val_split.set_format(type='torch', columns=['input_ids', 'attention_mask', 'labels'])
```

```python
from transformers import TrainingArguments , Trainer
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=5,
    per_device_train_batch_size=16,
    per_device_eval_batch_size=16,
    gradient_accumulation_steps=1,
    eval_strategy='steps',     
    eval_steps=500,
    save_strategy='steps',      
    save_steps=500,             
    lr_scheduler_type="linear",
    save_total_limit=2,
    load_best_model_at_end=True,
    metric_for_best_model="f1",
    learning_rate=1e-5,
    weight_decay=0.1,
    warmup_ratio=0.1,
    optim="adamw_torch",
    label_smoothing_factor=0.05,
    dataloader_num_workers=8,
    use_cpu=False,
    fp16=True,
    logging_steps=100,
)
```

```python
model = model.to('cuda')
print("Model device:", next(model.parameters()).device)
print("Device used by Trainer:", training_args.device)
```

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score

def compute_metrics(eval_pred):
    logits, labels = eval_pred
    predictions = np.argmax(logits, axis=-1)

    return {
        "accuracy": accuracy_score(labels, predictions),
        "f1": f1_score(labels, predictions, average="weighted")
    }
```

```python
from transformers import EarlyStoppingCallback
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_val_split["train"],
    eval_dataset=train_val_split["validation"],
    compute_metrics=compute_metrics,
)
```
